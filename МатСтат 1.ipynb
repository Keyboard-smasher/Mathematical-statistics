```python
import numpy as np
from scipy import stats
from matplotlib import pyplot
import seaborn as sns
```

```python
theta = 200
s = 100
x = np.random.randint(0, theta, size = s)

theta1 = [2/i*np.sum(x[:i]) for i in range (1, s+1)]
Theta1 = [np.abs(th - theta) for th in theta1]
theta2 = [np.min(x[:i]) * (i+1) for i in range (1, s+1)]
Theta2 = [np.abs(th - theta) for th in theta2]
theta3 = [np.max(x[:i]) * i/(i+2) for i in range (1, s+1)]
Theta3 = [np.abs(th - theta) for th in theta3]
theta4 = [np.max(x[:i]) + np.min(x[:i]) for i in range (1, s+1)]
Theta4 = [np.abs(th - theta) for th in theta4]
theta5 = []
theta5.append(x[0])
for i in range (2, s+1):
    theta5.append(x[0] + np.sum(x[1:i])*1/(i-1))
Theta5 = [np.abs(th - theta) for th in theta5]
```

```python
fig, axs = pyplot.subplots(1, 1, figsize=(13, 10))
    
axs.plot(Theta1, label = "\u03F41", color = "#0fff0f")
axs.plot(Theta2, label = "\u03F42", color = "#ff0fa0")
axs.plot(Theta3, label = "\u03F43", color = "red")
axs.plot(Theta4, label = "\u03F44", color = "brown")
axs.plot(Theta5, label = "\u03F45", color = "blue")
pyplot.legend(loc=1, prop={'size': 20})
pyplot.title("Theta distribution", fontsize = 20)
axs.set_xlabel("Amount of numbers", fontsize = 18)
axs.set_ylabel("|\u03F4i - \u03F4|", fontsize = 18)
pyplot.plot();
```
